Write a function to determine the mode of an array of integers.

The list may have one mode, more than one mode, or no mode at all. If the list has multiple modes, return them in ascending order.

Example 1:

Input:

nums = [1, 2, 2, 3, 4]
Output:

mode(nums) -> [2]
Example 2:

Input:

nums = [1,1,2,2]
Output:

mode(nums) -> [1, 2]

```python
from collections import Counter

def mode(nums):
    if not nums:
        return []

    freq = Counter(nums)
    max_freq = max(freq.values())

    # Get all numbers with max frequency
    modes = [num for num, count in freq.items() if count == max_freq]

    return sorted(modes)
```

You are given an array of strings arr, where each string is in the format: <letters><number>. For example, you might see "a1", "b32", or "caba522".

Write a function called custom_sort(arr), which sorts the list of strings in two ways: first by the letter, in alphabetical order, and second by the numbers in descending order.

Note: The letters can only be lower case, and have length 
≥

≥1.

Example:

Input:

arr = ["b12", "a15", "b7", "a3", "c4", "c20"]
Output:

custom_sort(arr) -> '["a15", "a3", "b12", "b7", "c20", "c4"]'

```python
import re

def custom_sort(arr):
    def sort_key(s):
        # Separate letters and numbers using regex
        match = re.match(r"([a-z]+)(\d+)", s)
        letters = match.group(1)
        number = int(match.group(2))
        return (letters, -number)  # sort letters ascending, numbers descending

    sorted_arr = sorted(arr, key=sort_key)
    return str(sorted_arr)
```

```python
def custom_sort(arr):
    """
    Sort a list of strings in the format <letters><number>:
    1. Primary sort: letters in alphabetical order
    2. Secondary sort: numbers in descending order

    Args:
        arr: List of strings in the format <letters><number>

    Returns:
        Sorted list of strings
    """
    # Parse each string into its letter and number components
    def parse_item(item):
        # Find the position where digits start
        i = 0
        while i < len(item) and not item[i].isdigit():
            i += 1

        # Split the string into letters and number
        letters = item[:i]
        number = int(item[i:]) if i < len(item) else 0

        return letters, number

    # Sort the array using a custom key function
    # Sort first by letters (ascending), then by numbers (descending)
    sorted_arr = sorted(arr, key=lambda x: (parse_item(x)[0], -parse_item(x)[1]))

    return sorted_arr
```

```python
from datetime import datetime
import numpy as np

def delta_buss_days(date1, date2):
    d1 = datetime.strptime(date1, "%Y-%m-%d").date()
    d2 = datetime.strptime(date2, "%Y-%m-%d").date()

    # Ensure d1 is earlier than d2
    start, end = sorted([d1, d2])

    # Count business days
    return np.busday_count(start.isoformat(), end.isoformat())

print(delta_buss_days("2021-01-31", "2021-02-18"))  # Output: 14
```

```python
# Compute mean, median, variance, and standard deviation of arr = np.array([1,2,3,4,5]).
arr = np.array([1,2,3,4,5])
mean_ = arr.mean()
median_ = np.median(arr)
var_ = arr.var(ddof=0)      # population variance
std_ = arr.std(ddof=0)      # population std

# sample stats (if asked): var_s = arr.var(ddof=1); std_s = arr.std(ddof=1)
mean_, median_, var_, std_
```

```python
x = np.array([3, 7, 1, 9, 5], dtype=float)
x_min, x_max = x.min(), x.max()
x_norm = (x - x_min) / (x_max - x_min)
x_norm
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

dot_prod = a @ b            # or np.dot(a,b) -> 32
elemwise = a * b            # -> array([ 4,10,18])
dot_prod, elemwise
```

```python
# Solve the linear system
# 2x + y = 5
# x − y = 1

A = np.array([[2,1],
              [1,-1]], dtype=float)
b = np.array([5,1], dtype=float)

sol = np.linalg.solve(A, b)  # [x, y]
x, y = sol
x, y
```

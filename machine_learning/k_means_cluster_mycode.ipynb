2. assign all points to closest centroid to form clausters
3. update centroid
4. repeat above untill 100 times
5. final assignment, return lables of input data

```python
import random
import numpy as np

def dist(a, b):
    return np.sqrt(sum((i-j)**2 for i, j in zip(a, b)))

def get_centroid(points):
    return list(np.mean(points, axis=0))

def kmeans(data, k, max_iters=100, tolerance=1e-4):
    centers = random.sample(data, k)
    
    for iteration in range(max_iters):
        groups = [[] for _ in range(k)]
        
        # Assign points to closest centers
        for point in data:
            distances = [dist(center, point) for center in centers]
            closest_idx = distances.index(min(distances))
            groups[closest_idx].append(point)
        
        # Calculate new centers
        new_centers = []
        for group in groups:
            if group:
                new_centers.append(get_centroid(group))
            else:
                new_centers.append(random.choice(data))
        
        # Check convergence
        center_shift = sum(dist(old, new) for old, new in zip(centers, new_centers))
        print(f'Iteration {iteration}: centers = {new_centers}, shift = {center_shift:.6f}')
        
        if center_shift < tolerance:
            print(f'Converged after {iteration + 1} iterations')
            break
            
        centers = new_centers
    
    # Final assignment
    labels = []
    for point in data:
        distances = [dist(center, point) for center in centers]
        labels.append(distances.index(min(distances)))
    
    return labels, centers


data = [
    [1, 2], [1.5, 1.8], [2, 2.2], [2.5, 2.0],
    [6, 8], [7, 9], [8, 8.5], [7.5, 7.8],
    [3, 6], [3.5, 5.5], [4, 6.5], [5, 5]
]
labels, centers = kmeans(data, k=3)
print("Labels:", labels)
print("Centers:", centers)
```

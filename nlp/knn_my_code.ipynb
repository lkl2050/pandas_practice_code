```python
from collections import Counter
import math

class knntextclassifier:
    def __init__(self, k=2):
        self.k = k
        self.train_data = []
        self.vocab = set()

    def preprocess_text(self, text):
        return text.lower().split()

    def vectorize(self, words):
        # dict1 = {}
        # for word in words:
        #     dict1[word] = words.count(word)

        # print('vectorize results=',dict1)
        # return dict1

        print('vectorize results=',Counter(words))
        return Counter(words)

    def build_vocab(self, train_x):
        for x in train_x:
            procx = self.preprocess_text(x)
            self.vocab.update(procx)

    def fit(self, train_x, train_y):
        self.build_vocab(train_x)
        for x, y in zip(train_x, train_y):
            procx = self.preprocess_text(x)
            vercx = self.vectorize(procx)
            self.train_data.append((vercx, y))


    def compute_distance(self, vec1, vec2):
        res= 0
        for word in self.vocab:
            diff = (vec1.get(word, 0) - vec2.get(word, 0))**2
            res+= diff
        return res

    def predict(self, test_x):
        pred_res = []

        for x in test_x:
            procx = self.preprocess_text(x)
            test_vec = self.vectorize(procx)
            distances = []
            for train_vec, label in self.train_data:
                distance = self.compute_distance(test_vec, train_vec)
                distances.append((distance, label))
            distances = sorted(distances, key = lambda x: x[0])

            print('distances', distances)


            k_nearest_labels = []
            for i in range(self.k):
                k_nearest_labels.append(distances[i][1])

            print('k_nearest_labels', k_nearest_labels)
            most_common_label = Counter(k_nearest_labels)

            res = max(most_common_label, key = most_common_label.get)

            pred_res.append(res)
        return pred_res


train_x = ['this is kk techology','this is kk techology 2', 'pet pet pet ']
train_y = ['tech', 'tech', 'pet']
test_x = ['kk techology ','pet pet pet']

knn_instance = knntextclassifier(2)
knn_instance.fit(train_x, train_y)
knn_instance.predict(test_x)
```

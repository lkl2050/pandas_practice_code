```python
from collections import Counter

def build_bow(texts):
    # Split words and lowercase
    tokenized_texts = [text.lower().split() for text in texts]
    
    # Build vocabulary
    all_words = {word for doc in tokenized_texts for word in doc}

    # Step 3: build a sorted vocabulary list
    vocab = sorted(all_words)
    
    # Convert each text to a vector
    vectors = []
    for text in tokenized_texts:
        counts = Counter(text)
        vector = [counts.get(word, 0) for word in vocab]
        vectors.append(vector)
    
    return tokenized_texts, vocab, vectors

texts = ["I love AI", "AI is amazing", "I love machine learning"]
tokenized_texts, vocab, vectors = build_bow(texts)
print("tokenized_texts:", tokenized_texts)
print("Vocabulary:", vocab)
print("Vectors:", vectors)
```

```python
from collections import Counter
def build_bow(texts):
    tokenized_texts = []
    for i in texts:
        tokenized_texts.append( i.lower().split() )
    
    vocab = set()
    for i in tokenized_texts:
        for w in i:
            vocab.add(w)
    
    vectors = []
    for text in tokenized_texts:
        counts = Counter(text)
        vector = [counts.get(word, 0) for word in vocab]
        vectors.append(vector)
            


    return tokenized_texts, vocab, vectors





texts = ["I love AI AI AI", "AI is amazing", "I love machine learning"]
tokenized_texts, vocab, vectors = build_bow(texts)
print("tokenized_texts:", tokenized_texts)
print("Vocabulary:", vocab)
print("Vectors:", vectors)
```

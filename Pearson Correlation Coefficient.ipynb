Given two vectors, represented as lists X and Y, return the Pearson Correlation Coefficient.

```python
import math

def pearson_correlation(X, Y):
    n = len(X)
    if n != len(Y) or n == 0:
        raise ValueError("Vectors must be of equal non-zero length.")

    mean_x = sum(X) / n
    mean_y = sum(Y) / n

    num = sum((x - mean_x) * (y - mean_y) for x, y in zip(X, Y))
    denom_x = sum((x - mean_x) ** 2 for x in X) ** 0.5
    denom_y = sum((y - mean_y) ** 2 for y in Y) ** 0.5

    if denom_x == 0 or denom_y == 0:
        return 0  # or float('nan'), depending on your preference

    return num / (denom_x * denom_y)
    
```

```python
import numpy as np

def pearson_correlation(X, Y):
    X = np.array(X)
    Y = np.array(Y)
    
    if X.shape != Y.shape or X.size == 0:
        raise ValueError("Vectors must be of equal non-zero length.")

    mean_x = X.mean()
    mean_y = Y.mean()

    num = np.sum((X - mean_x) * (Y - mean_y))
    denom = np.sqrt(np.sum((X - mean_x) ** 2)) * np.sqrt(np.sum((Y - mean_y) ** 2))

    if denom == 0:
        return 0  # or np.nan

    return num / denom
```
